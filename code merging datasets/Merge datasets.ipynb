```python
import pandas as pd
import numpy as np
import re
```

```python
# importer les datasets a merge
df_nb_heureTravail=pd.read_csv("../datasets_dw/Racha/df_nb_heureTravail_primaire_college.csv", sep=';')
df_nb_semaines_conge=pd.read_csv("../datasets_dw/Racha/df_nb_semaines_conge.csv",  sep=';')
df_imigration=pd.read_csv("../datasets_dw/Racha/immigration-by-country-2024.csv")
df_political_regime=pd.read_csv("../datasets_dw/Racha/political-regime.csv")

df_nbPrix_musees_anneeSouv=pd.read_csv("../datasets_scraped/df_nbPrix_musees_anneeSouv_univTop500.csv")
```

```python
def has_common_name(liste_mots1, liste_mots2):
    
    liste_mots_communs=set()

    # Vérifier si on a au moins un mot en commun
    for mot in liste_mots1:
        if mot in liste_mots2:
            liste_mots_communs.add(mot)
    return liste_mots_communs
```

## Merge dataset polical_regime

```python
# Recuperer les pays et leur régime politique pour l'année 2019
df_political_regime_2019=df_political_regime.loc[df_political_regime.Year==2019, ['Entity','Political regime']]
df_political_regime_2019.rename(columns={"Entity":"Pays"}, inplace=True)
df_political_regime_2019
```

```python
df_nbPrix_musees_anneeSouv_PR=df_nbPrix_musees_anneeSouv.set_index('Pays').join(df_political_regime_2019.set_index('Pays'))
df_nbPrix_musees_anneeSouv_PR_nan=df_nbPrix_musees_anneeSouv_PR[df_nbPrix_musees_anneeSouv_PR['Political regime'].isna()]
print("Le nombre de pays sans l'information sur le régime politique : ",np.sum(df_nbPrix_musees_anneeSouv_PR_nan['Political regime'].isna()))
df_nbPrix_musees_anneeSouv_PR_nan
```

```python
# Identifier les mots en communs entre le dataset que nous voulons join et celui des pays sans correspondance (retrouver les pays mal etiquetés)
df_nbPrix_musees_anneeSouv_PR_nan.reset_index(inplace=True)
liste_pays1=list(df_nbPrix_musees_anneeSouv_PR_nan.Pays.apply(lambda x : x.split()))
liste_pays1=[mot for sublist in liste_pays1 for mot in sublist]

liste_pays2=list(df_political_regime_2019.Pays.apply(lambda x : x.split()))
liste_pays2=[mot for sublist in liste_pays2 for mot in sublist]

liste_mots_communs=has_common_name(liste_pays1,liste_pays2)
liste_mots_communs 
```

```python
# Mettre a jour le nom des pays de sorte à ce qu'ils correspondent à ceux du dataset musée
dict_convertNomPays={"Palestine/Gaza":"Palestine"}
df_political_regime_2019.Pays=df_political_regime_2019.Pays.apply(lambda x : dict_convertNomPays[x] if x in dict_convertNomPays else x)
df_political_regime_2019[df_political_regime_2019.Pays=='Palestine']
```

```python
#df_nbPrix_musees_anneeSouv.reset_index(inplace=True)
df_nbPrix_musees_anneeSouv_PR_final=df_nbPrix_musees_anneeSouv.set_index('Pays').join(df_political_regime_2019.set_index('Pays'))
print("Le nombre de pays sans l'information sur le régime politique : ",np.sum(df_nbPrix_musees_anneeSouv_PR_final['Political regime'].isna()))
```

```python
df_merged1=df_nbPrix_musees_anneeSouv_PR_final.copy()
df_merged1
```

```python
df_nbPrix_musees_anneeSouv_PR_final.reset_index(inplace=True)
df_nbPrix_musees_anneeSouv_PR_final.to_csv("../datasets_merged/df_nbPrix_musees_anneeSouv_PR.csv", index=False)
```

## Merge dataset heureTravail

```python
df_merged2=pd.read_csv("../datasets_merged/df_nbPrix_musees_anneeSouv_PR_imig.csv")
```

```python
df_nb_heureTravail.rename(columns={"Category":'Pays'}, inplace=True)
df_nb_heureTravail=df_nb_heureTravail.loc[:,['Pays','Primary','Lower secondary']]
df_nb_heureTravail.Pays=df_nb_heureTravail.Pays.apply(lambda x : re.sub(r'[^a-zA-Z ]', '', x))
df_nb_heureTravail.Pays=df_nb_heureTravail.Pays.apply(lambda x : x[:-1])

renamePays={"Trkiye":"Turkey", "Slovak Republic":"Slovakia", 'French Comm Belgium':"Belgium", "Flemish Comm Belgium":"Belgium" }
df_nb_heureTravail.Pays=df_nb_heureTravail.Pays.apply(lambda x : renamePays[x] if x in renamePays else x)
df_nb_heureTravail
```

```python
# Corriger les doublons dans certains pays (Belgium était réparti en deux régions)
df_nb_heureTravail=df_nb_heureTravail.groupby('Pays').mean()
df_nb_heureTravail.reset_index(inplace=True)
df_nb_heureTravail
```

```python
df_merged2_nb_heureTravail=df_merged2.set_index('Pays').join(df_nb_heureTravail.set_index('Pays'))
df_merged2_nb_heureTravail_nan=df_merged2_nb_heureTravail[df_merged2_nb_heureTravail['Primary'].isna()]
print("Le nombre de pays sans l'information sur le nombre d'heures de travail des étaudiants de primaire et collège : ",str(np.sum(df_merged2_nb_heureTravail_nan['Primary'].isna())-3))
df_merged2_nb_heureTravail_nan
```

## Merge dataset semainesCongé

```python
df_nb_semaines_conge.rename(columns={"Category":'Pays'}, inplace=True)
df_nb_semaines_conge.Pays=df_nb_semaines_conge.Pays.apply(lambda x : re.sub(r'[^a-zA-Z ]', '', x))

renamePays={"Trkiye":"Turkey", "Slovak Republic":"Slovakia", 
            'French Comm Belgium':"Belgium", "Flemish Comm Belgium":"Belgium" ,
            "England UK":"UK", "ScotlandUK":"UK"}
df_nb_semaines_conge.Pays=df_nb_semaines_conge.Pays.apply(lambda x : renamePays[x] if x in renamePays else x)
df_nb_semaines_conge
```

```python
df_nb_semaines_conge.iloc[:,1:]=df_nb_semaines_conge.iloc[:,1:].applymap(lambda x: float(re.sub(r',', '.', x)) if isinstance(x, str) else x)
df_nb_semaines_conge
```

```python
# Corriger les doublons dans certains pays (Belgium et UK étaient répartis en deux régions)
df_nb_semaines_conge=df_nb_semaines_conge.groupby('Pays').mean()
df_nb_semaines_conge.reset_index(inplace=True)
df_nb_semaines_conge
```

```python
df_merged2_nb_heureTravail_nb_semaines_conge=df_merged2_nb_heureTravail.join(df_nb_semaines_conge.set_index('Pays'))
df_merged2_nb_heureTravail_nb_semaines_conge.reset_index(inplace=True)
df_merged2_nb_heureTravail_nb_semaines_conge
```

```python
df_merged2_nb_heureTravail_nb_semaines_conge.to_csv("../datasets_merged/df_mergedPart1.csv",index=False)
```
