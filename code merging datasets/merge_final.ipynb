```python
import pandas as pd
```

```python
df=pd.read_csv('../datasets_merged/df_mergedPart1.csv')
```

```python
df=df.rename(columns={'Pays':'country'})
df.loc[df['country']=='Republic of the Congo','Annee_souverainete']=1884.0
df.loc[(df['country']=='Congo') & (df['Annee_souverainete']==1960.0),'country']='DR Congo'
df
```

```python
df[(df['country']=='Congo') & (df['Annee_souverainete']==1884.0)]
df=df.drop([42])
```

```python
df_part2=pd.read_csv('../datasets_merged/df_mergedPart2.csv')
```

```python
df_part2
```

```python
df_merge=pd.merge(df,df_part2,on='country',how='outer')
df_merge
```

```python
df_iq=pd.read_csv("../data/IQ_pays.csv")
df_iq
```

```python
df_merged_final=pd.merge(df_merge,df_iq,on='country',how='outer')
df_merged_final
```

```python
df_merged_final[df_merged_final['nb_habitants'].notnull()]
```

```python
df_merged_final['GDP_perCapita']=df_merged_final['GDP']/df_merged_final['nb_habitants']
```

```python
df_merged_final['GDP_perCapita']
```

```python
df_merged_final['Indice_evolution']=df_merged_final['GDP_perCapita']/(2024-df_merged_final['Annee_souverainete']) 
```

```python
df_merged_final[df_merged_final['country']=='Singapore']['Indice_evolution'] #Indice elevÃ© pour le Singapour
```

```python
df_merged_final.iloc[df_merged_final['Indice_evolution'].argmax()] #le pays ayant l'indice le plus Ã©levÃ© est le Quatar
```

```python
df_merged_final.columns
```

```python
modif={}
for column in df_merged_final.columns:
    modif[column] = column.title().replace(" ","_").replace("-","_")

modif['Superficie']="Area"
modif['nb_diplomes']="Nb_graduates"
modif['nb_habitants']="Population"
modif['Rang']="Mean_Rank_Univ"
modif['Global rank']="Global_rank_Literacy_rate"

modif
```

```python
df_merged_final.rename(columns = modif)
```

```python
df_merged_final.to_csv("../datasets_merged/df_mergedFinal.csv",index=False)
```
