```python
res={
    'pays':set(),
    'country':set()
}
out={}
nb=0
```

```python
for cnt in df['Pays']:
    out[cnt]=False
for cnt in df['Pays']:
    for val in iq["country"]:
        if out[cnt]==False and cnt!=val and ((cnt in val) or (val in cnt)):
            if cnt!='Niger':
                df[df['Pays']==cnt]=val
            print(cnt,"=>",val)
        if cnt==val: 
            out[cnt]=True
            nb+=1
print(nb)
```
