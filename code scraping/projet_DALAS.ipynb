```python
import requests
import pandas as pd
from scrapy.selector import Selector
import pandas as pd
import re
```

```python
def transform_key(key):
    pattern = r'(?<=of-)([a-zA-Z]+(-[a-zA-Z]+)?)(?=-)'
    result = re.search(pattern, key)
    if result:
        return result.group(1)
    else: return key

print(transform_key("countries-of-north-america-23"))
```

```python
url='https://www.clearias.com/countries-world-listed-continent/'
reponse =requests.get(url).text
```

```python
data = {}
response=Selector(text=str(reponse))
current_title = None
edited=None
for element in response.xpath('//h3|//ol'):
  if element.xpath('name()').get() == 'h3':
    current_title = element.xpath('span/@id').get()
    if current_title:
      edited=transform_key(current_title)
      data[edited] = []
  elif element.xpath('name()').get() == 'ol' and current_title:
    items = element.xpath('li/text()').getall()
    if items:
      data[edited].extend(items)
      current_title=None
```

```python
data
```

```python
for element in response.xpath('//div[@class="pf-content"]/h2[2]'):
  title_id = element.xpath('span/@id').get()
  if title_id:
    data[title_id] = []
    for ol_element in element.xpath('following-sibling::ol[1]'):
      items = ol_element.xpath('li/text()').getall()
      data[title_id].extend(items)
```

```python
data
```

```python
data_tuples = [(key, value) for key, values in data.items() for value in values]
```

```python
df = pd.DataFrame(data_tuples, columns=['Continent', 'Pays'])

print(df)
```

```python
df.to_csv('continent.csv', index=False)

print("DataFrame saved to 'continent.csv'")
```

# Superficie

```python
url='https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_area'
reponse =requests.get(url).text
```

```python
response=Selector(text=str(reponse))
```

```python
table_rows = response.xpath('//div/table[contains(@class, "wikitable") and contains(@class, "sortable") and contains(@style,"text-align:right")]/tbody//tr')
superficie = []
print(len(table_rows))
for row in table_rows[2:]:

  col_2 = row.xpath('td[2]/a/@title').get()
  if col_2 is None:
    col_2 = row.xpath('td[2]/i/a/@title').get()
    if col_2 is None:
      col_2 = row.xpath('td[1]/a/@title').get()
      col_3=row.xpath('td[2]//text()').get()
      superficie.append({'Pays': col_2.strip(), 'Superficie': col_3.strip()})
      continue
  print(col_2)
  col_3 = row.xpath('td[3]//text()').get()
  if col_2 is not None and col_3 is not None:
            superficie.append({'Pays': col_2.strip(), 'Superficie': col_3.strip()})
print(superficie)
```

```python
ds = pd.DataFrame(superficie)
ds
```

```python
def extract_first_number(string):
    pattern = r'([\d.,]+)'
    matches = re.findall(pattern, string)
    if matches:
        first_number = matches[0]
        first_number = first_number.replace(',', '').replace('.', ',')
        return first_number
    else:
        return string

ds['Superficie_edited'] = ds['Superficie'].apply(lambda x: extract_first_number(x))
```

```python
ds.to_csv('superficie.csv', index=False)

print("DataFrame saved to 'superficie.csv'")
```
