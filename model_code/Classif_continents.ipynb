```python
import pandas as pd
import numpy as np
pd.DataFrame.iteritems = pd.DataFrame.items
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("../datasets_merged/df_mergedFinal.csv")
df
```

```python
numerical_columns = df.select_dtypes(include=['number'])
ds=numerical_columns.drop(columns=['Annee_Souverainete','Break_1','Break_2', 'Break_3', 'Break_4', 'Break_5','Political_Regime','Mean_Rank_Univ','Averageiq_Ici2017Score'])
ds
```

```python
#garder que les variables dont au moins 100 valeurs sont renseignées:

colonnes=list(ds.describe().columns[ds.describe().loc['count'] > 100])
colonnes
```

```python
ds=ds[colonnes].dropna()
ds
```

```python
X = ds # Sélectionner toutes les colonnes 
y = df.loc[ds.index,'Continent']  # Sélectionner uniquement la colonne des labels
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialiser l'encodeur de labels
label_encoder = LabelEncoder()

# Adapter l'encodeur aux données de labels et le transformer en entiers
labels = label_encoder.fit_transform(y)

# Afficher les labels encodés
print("Labels encodés :", labels)
```

# SVM

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Diviser les données en ensembles d'entraînement et de test
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

# Initialiser le modèle SVM
svm_model = SVC()

# Entraîner le modèle SVM
svm_model.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
svm_predictions = svm_model.predict(X_test)

# Calculer l'exactitude du modèle
svm_accuracy = accuracy_score(y_test, svm_predictions)
print("Accuracy SVM :", svm_accuracy)
```

# Reg logistique

```python
from sklearn.linear_model import LogisticRegression

# Initialiser le modèle de régression logistique avec la pondération des classes
lr_model_weighted = LogisticRegression(class_weight='balanced')

# Entraîner le modèle de régression logistique
lr_model_weighted.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
lr_predictions_weighted = lr_model_weighted.predict(X_test)

# Calculer l'accuracy du modèle
lr_accuracy_weighted = accuracy_score(y_test, lr_predictions_weighted)
print("Accuracy Régression Logistique avec pondération des classes :", lr_accuracy_weighted)
```

# XGboost

```python
import xgboost as xgb

# Initialiser le modèle XGBoost avec la pondération des classes
xgb_model_weighted = xgb.XGBClassifier(scale_pos_weight=len(y_train[y_train==0])/len(y_train[y_train==1]))

# Entraîner le modèle XGBoost
xgb_model_weighted.fit(X_train, y_train)

# Faire des prédictions sur l'ensemble de test
xgb_predictions_weighted = xgb_model_weighted.predict(X_test)

# Calculer l'exactitude du modèle
xgb_accuracy_weighted = accuracy_score(y_test, xgb_predictions_weighted)
print("Exactitude XGBoost avec pondération des classes :", xgb_accuracy_weighted)
```
